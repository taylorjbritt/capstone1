```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt


def unpickle(filename):
    df = pd.read_pickle('../data/pickles/' + filename + '.pkl')
    return df

denver = unpickle('denver')

paris = unpickle('paris')

saopaulo = unpickle('saopaulo')

santiago = unpickle('santiago')

nyc = unpickle('nyc')

la = unpickle('la')




```

```python
def plot_concentrations(df, name):
    fig = plt.figure(figsize=(12,12))
    ax = fig.add_subplot(2,1,1)
    ax.scatter(df['date'], df['pm25'], label = "PM2.5" )
    ax.set_title(name +" "+ 'Air Quality - PM2.5')
    ax.set_ylim(0, (df['pm25'].max()*1.1))
    ax.legend(loc='best')
    fig.tight_layout(pad=1)
    ax = fig.add_subplot(2,1,2)
    ax.scatter(df['date'], df['no2'], label = "NO2" )
    ax.set_title(name +" "+ 'Air Quality - NO2')
    ax.set_ylim(0, (df['no2'].max()*1.1))
    ax.legend(loc='best')
    fig.tight_layout(pad=1)
```

```python
plot_concentrations(saopaulo, 'So Paulo')
```

```python
plot_concentrations(santiago, 'Santiago de Chile')
```

```python
plot_concentrations(denver, 'Denver')
```

```python
plot_concentrations(nyc, 'New York City')
```
